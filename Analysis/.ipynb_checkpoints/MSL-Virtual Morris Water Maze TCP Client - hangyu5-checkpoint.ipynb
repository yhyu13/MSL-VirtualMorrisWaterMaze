```python
'''
    Set up some parsing functions first.
'''
import io
import cv2

# Helper function for decoding state information
def decode_state(state_message):
    raise NotImplementedError
    return None

# Helper function for parsing messages
def parse_message(message):
    # For image messages
    if message[0:len("Music")] == bytes("Music", 'utf8'):
        # If we receive state information, decode it
        return {'type':"Music", 
                'value': 100}
    elif message[0:len("Image")] == bytes("Image", 'utf8'):
        # There may be a cleaner way to do this conversion, but this works fast enough for comfort
        return {'type':"Image", 
                'value': plt.imread(io.BytesIO(message[len("Image"):]),format='JPG')}
    elif message[0:len("State")] == bytes("State", 'utf8'):
        # If we receive state information, decode it
        return {'type':"State", 
                'value': decode_state(message[len("State"):-1])}
    else:
        # Default messages are returned as-is (with type as None and value as message)
        return {'type': None, 
                'value': message}
```

```python
# From A3CDoom

# Copies one set of variables to another.
# Used to set worker network parameters to those of global network.
def update_target_graph(from_scope,to_scope):
    from_vars = tf.get_collection(tf.GraphKeys.TRAINABLE_VARIABLES, from_scope)
    to_vars = tf.get_collection(tf.GraphKeys.TRAINABLE_VARIABLES, to_scope)

    op_holder = []
    for from_var,to_var in zip(from_vars,to_vars):
        op_holder.append(to_var.assign(from_var))
    return op_holder

# Discounting function used to calculate discounted returns.
def discount(x, gamma):
    return scipy.signal.lfilter([1], [1, -gamma], x[::-1], axis=0)[::-1]

#Used to initialize weights for policy and value output layers
def normalized_columns_initializer(std=1.0):
    def _initializer(shape, dtype=None, partition_info=None):
        out = np.random.randn(*shape).astype(np.float32)
        out *= std / np.sqrt(np.square(out).sum(axis=0, keepdims=True))
        return tf.constant(out)
    return _initializer
```

```python
# Modified from A3CDoom to fit RGB images

# Processes Doom screen image to produce cropped and resized image.
def process_frame(s,shape=[84,84],normalized=True):
    s = scipy.misc.imresize(s,shape)
    r,g,b = cv2.split(s)
    s = np.array([r,g,b]).transpose(1,2,0)
    if normalized:
        s = s / 255.0
    return s
```

# Game Start

```python
'''
    Set up the communication client and begin. Keyboard Interrupt closes.
'''

import socket
import select
import numpy as np
import time
import cv2

# Plot inline
%matplotlib inline
# Set up OpenCV Window
cv2.startWindowThread()

# TCP Socket Parameters
TCP_IP = '192.168.56.1'
TCP_PORT = 5005
# Data Buffer
BUFFER_SIZE = 1024
data = b''

# End Message Token
END_TOKEN = "<EOF>"

# Set up Socket
s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
s.connect((TCP_IP, TCP_PORT))
s.setblocking(0)

# Set Start Trial Option
s.send(bytes("PlayerPrefsStrial,Practice - Hills" + END_TOKEN, 'utf8'))

# Start Scene
s.send(bytes("Scene1" + END_TOKEN, 'utf8'))

# Request First Image Frame
# s.send(bytes("ImageRequest" + END_TOKEN, 'utf8'))

# hangyu5:
action = ['KeyUP','KeyDOWN','KeyLEFT','KeyRIGHT'] # What are the key bindings? And how could agent look around?
memory_size = 100 # 
images = [None] * memory_size
itick = 0
gif_count = 0
time1 = time.time()
# hangyu5.

# Loop until keyboard interrupt, reading/requesting data
while True:
    try:
        # Wait for data and store in buffer
        ready = select.select([s], [], [], 0)
        if ready[0]:
            try:
                data += s.recv(BUFFER_SIZE)
            except ConnectionResetError:
                print("The connection was closed.")
                break
        
        # Check if buffer has a full message
        if(bytes(END_TOKEN, 'utf8') in data):
            # Decode the message and clear it from the buffer
            idx = data.index(bytes(END_TOKEN, 'utf8'))
            while idx != -1:
                parsed_message = parse_message(data[0:idx])
            
                # If the message was an image, display it using OpenCV (faster than matplotlib)
                if parsed_message['type'] == 'Image':
                    
                    
                    screen_BGR = parsed_message['value']
                    screen_RGB = cv2.cvtColor(screen_BGR, cv2.COLOR_BGR2RGB)
                    cv2.imshow('frame', screen_RGB)
                    cv2.waitKey(1)
                    
                    
                    
                    # Send a request for a new frame
                    
                    s.send(bytes('KeyUP' + END_TOKEN, 'utf8'))
                    # s.send(bytes("ImageRequest" + END_TOKEN, 'utf8'))
                    
                    # hangyu5: 
                    
                    images[itick] = process_frame(screen_BGR,normalized=False)
                    
                    itick += 1
                    
                    if itick == memory_size:
                        
                        # count duration
                        time2 = time.time()
                        duration = time2 - time1
                        print(time2-time1)
                        
                        # save episodes as .gif
                        gif = np.array(images)
                        make_gif(gif,r'./memory/gif_' + str(gif_count)+'.gif',
                            duration=len(gif)*.0625,true_image=True,salience=False)
                        gif_count += 1
                        
                        # reset variables
                        itick = 0
                        images = [None] * memory_size
                        time1 = time.time()
                    # hangyu5.

                data = data[idx + len(END_TOKEN):-1]
                if(data != ''):
                    try:
                        idx = str(data).index(END_TOKEN)
                    except ValueError:
                            break

    # Wait for a keyboard interrupt to stop the loop
    except KeyboardInterrupt:
        break

# Close the socket and destroy the display window
s.close()
cv2.destroyAllWindows()

# Confirm exit
print('Done')
```
